# Toxic Comment Classification


```python

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report
```

Load and Explore the Data


```python
df = pd.read_csv('./data/train.csv')
df.head(10)
```

```python
print(df.info() , "/n")
df.isnull().sum()
```

Data Preprocessing


```python
X = df['comment_text'].fillna('')
Y = df['toxic'].astype('int')

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(X.shape, X_train.shape, X_test.shape)
```

Feature Extraction


```python
vectorizer = TfidfVectorizer(min_df=1, stop_words='english')

X_train_features = vectorizer.fit_transform(X_train)
X_test_features = vectorizer.transform(X_test)
```

Train the Model (Logistic Regression)


```python
model = LogisticRegression()
model.fit(X_train_features, Y_train)
```

```python
X_train_prediction = model.predict(X_train_features)
training_accuracy = accuracy_score(Y_train, X_train_prediction)
print(f'Accuracy on training data : {training_accuracy}')
```

```python
X_test_prediction = model.predict(X_test_features)
test_accuracy = accuracy_score(Y_test, X_test_prediction)
print(f'Accuracy on test data : {test_accuracy}\n')
print(classification_report(Y_test, X_test_prediction))
```

Test by taking input from the user and predicting


```python
sample_comment = ["You are a great person, thank you so much for your help!"]

sample_features = vectorizer.transform(sample_comment)
prediction = model.predict(sample_features)
print(prediction)

if prediction[0] == 1:
    print('The comment is Toxic.')
else:
    print('The comment is Clean (not toxic).')
```
